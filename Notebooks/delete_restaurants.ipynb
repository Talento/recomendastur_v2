```python
import pandas as pd

df_restaurants = pd.read_csv("../Scripts/Definitive_data/restaurant_processed.csv")

list_delete = ["Kebab", "Burger King", "Domino's", "Cepsa", "Repsol", "McDonald's", "Papa Johns", "Pizza Móvil", "Popeyes", "Telepizza", "KFC", "VIPS"]

list = "|".join(list_delete)  # crea la expresión 'Bar Pepe|Sidrería Juan|Casa María'

df = df_restaurants[~df_restaurants["Nombre"].str.contains(list, case=False, na=False)]

df.to_csv("../Scripts/Definitive_data/restaurant_processed_cleaned.csv", index=False)
```
